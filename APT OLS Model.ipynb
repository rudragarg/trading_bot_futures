```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def get_betas(column):
    df['bins'] = pd.qcut(df[column], 10, labels = ['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th', '10th'])
    df['bins'].value_counts().sort_index()

    y_bins = df.groupby('bins')['ytd_list'].mean()
    x = df.groupby('bins')[['pegRatio_list','priceToBook','moving_average_200','dividend_yield','currentRatio_list']].mean()
    y_10 = y_bins[9:10]
    y_1 = y_bins[:1]

    y = y_1[0]-y_10[0]

    bin_1 = df[df["bins"] == "1st"]
    bin_1["range_peg"] = bin_1["pegRatio_list"].apply(lambda num: x.loc["10th", "pegRatio_list"] - num)
    bin_1["range_priceToBook"] = bin_1["priceToBook"].apply(lambda num: x.loc["10th", "priceToBook"] - num)
    bin_1["range_moving_average_200"] = bin_1["moving_average_200"].apply(lambda num: x.loc["10th", "moving_average_200"] - num)
    bin_1["range_dividend_yield"] = bin_1["dividend_yield"].apply(lambda num: x.loc["10th", "dividend_yield"] - num)
    bin_1["range_currentRatio_list"] = bin_1["currentRatio_list"].apply(lambda num: x.loc["10th", "currentRatio_list"] - num)
    bin_1['range_ytd_list'] = bin_1["ytd_list"].apply(lambda num: y_bins.iloc[9] - num)

    bin_1[["range_peg"]] = bin_1[["range_peg"]].apply(lambda x: (x - x.min()) / (x.max() - x.min())) 

    X = bin_1[["range_peg", "range_priceToBook", "range_moving_average_200", "range_dividend_yield", "range_currentRatio_list"]]
    y = bin_1["range_ytd_list"]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    reg = LinearRegression().fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    y_pred = reg.predict(X_test)
    print("MSE:", mean_squared_error(y_test, y_pred))
    print("Betas:", reg.coef_)
    print("Intercept:", reg.intercept_)

    
```
