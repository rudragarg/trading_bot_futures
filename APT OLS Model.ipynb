```python
apt = pd.DataFrame()
stock_list=[]
ytd_list=[]
pegRatio_list=[]
currentRatio_list = []
book_value=[]
dividend_yield=[]
moving_average_200 = []
priceToBook = []
for i in symbols:
    stock = yf.Ticker(i)
    stock1 = stock.info
    if stock1['regularMarketPreviousClose'] != None:
        pegRatio = stock1['pegRatio']
        history = stock.history(period='1y')
        last_year = history['Close'][0]
        today= history['Close'][-1]
        ytd = (today-last_year)/last_year
        ytd_list.append(ytd)
        pegRatio_list.append(pegRatio)
        stock_list.append(i)
        currentRatio_list.append(stock1['currentRatio'])
        book_value.append(stock1['bookValue'])
        dividend_yield.append(stock1['fiveYearAvgDividendYield'])
        moving_average_200.append(stock1['twoHundredDayAverage'])
        priceToBook.append(stock1['priceToBook'])
        print(i)
    else:
        continue
```

```python
df = pd.DataFrame()
```
